```python

plt.figure(figsize=(7, 6))


plt.subplot(2, 2, 1)
df["State"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, cmap="Set3", wedgeprops={'edgecolor': 'black'})
plt.title("State-wise Distribution")


plt.subplot(2, 2, 2)
df["Aadhaar Linked"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=["lightblue", "lightcoral"], wedgeprops={'edgecolor': 'black'})
plt.title("Aadhaar Linked Status")


plt.subplot(2, 2, 3)
df["Gender"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=["gold", "lightgreen", "lightpink"], wedgeprops={'edgecolor': 'black'})
plt.title("Gender Distribution")


plt.subplot(2, 2, 4)
df["Foodgrain Type"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, cmap="coolwarm", wedgeprops={'edgecolor': 'black'})
plt.title("Foodgrain Type Distribution")


plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 9))


plt.subplot(2, 2, 1)
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")


plt.subplot(2, 2, 2)
sns.histplot(df["Monthly Quota (kg)"], bins=20, kde=True, color="orange")
plt.title("Monthly Quota (kg) Distribution")


plt.subplot(2, 2, 3)
sns.histplot(df["Subsidy Amount (INR)"], bins=20, kde=True, color="green")
plt.title("Subsidy Amount (INR) Distribution")


plt.show()
```

```python

plt.figure(figsize=(7, 10))

# Bar graph for Top 10 Districts by Transactions
plt.subplot(3, 1, 1)
df["District no."].value_counts().head(10).plot(kind="bar", color="purple", edgecolor="black")
plt.title("Top 10 Districts by Transactions")
plt.xlabel("Districts")
plt.ylabel("Transaction Count")

# Bar graph for FPS ID usage frequency
plt.subplot(3, 1, 2)
df["FPS ID"].value_counts().head(10).plot(kind="bar", color="blue", edgecolor="black")
plt.title("Top 10 Fair Price Shops by Usage")
plt.xlabel("FPS ID")
plt.ylabel("Number of Transactions")

# Bar graph for State-wise Success Transactions
plt.subplot(3, 1, 3)
success_state = df[df["Transaction Status"] == "Success"]["State"].value_counts()
success_state.plot(kind="bar", color="green", edgecolor="black")
plt.title("State-wise Successful Transactions")
plt.xlabel("State")
plt.ylabel("Success Count")

# Show bar graphs
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 8))

# Box plot for Subsidy Amount
plt.subplot(2, 2, 1)
sns.boxplot(y=df["Subsidy Amount (INR)"], color="orange")
plt.title("Box Plot of Subsidy Amount")

# Box plot for Monthly Quota
plt.subplot(2, 2, 2)
sns.boxplot(y=df["Monthly Quota (kg)"], color="blue")
plt.title("Box Plot of Monthly Quota (kg)")

# Box plot for Family Size
plt.subplot(2, 2, 3)
sns.boxplot(y=df["Family Size"], color="green")
plt.title("Box Plot of Family Size")

# Show box plots
plt.tight_layout()
plt.show()
```

```python



plt.figure(figsize=(16, 12))

# Top 15 States by Transactions
plt.subplot(3, 2, 1)
df["State"].value_counts().head(15).plot(kind="bar", color="teal", edgecolor="black")
plt.title("Top 15 States by Transactions")
plt.xlabel("State")
plt.ylabel("Transaction Count")

# Top 15 Districts by Transactions
plt.subplot(3, 2, 2)
df["District no."].value_counts().head(15).plot(kind="bar", color="purple", edgecolor="black")
plt.title("Top 15 Districts by Transactions")
plt.xlabel("Districts")
plt.ylabel("Transaction Count")

# Aadhaar Linked Status by State (Increased Visibility)
plt.subplot(3, 2, 3)
aadhaar_state = df.groupby("State")["Aadhaar Linked"].value_counts().unstack()
aadhaar_state.plot(kind="bar", stacked=True, colormap="coolwarm", edgecolor="black", ax=plt.gca())
plt.title("State-wise Aadhaar Linked Status")
plt.xlabel("State")
plt.ylabel("Count")
plt.legend(title="Aadhaar Linked")

# Success vs Failed Transactions by State
plt.subplot(3, 2, 4)
success_state = df.groupby("State")["Transaction Status"].value_counts().unstack()
success_state.plot(kind="bar", stacked=True, colormap="viridis", edgecolor="black", ax=plt.gca())
plt.title("State-wise Success vs Failed Transactions")
plt.xlabel("State")
plt.ylabel("Count")
plt.legend(title="Transaction Status")

# Top 15 FPS (Fair Price Shops) by Usage
plt.subplot(3, 2, 5)
df["FPS ID"].value_counts().head(15).plot(kind="bar", color="orange", edgecolor="black")
plt.title("Top 15 FPS by Usage")
plt.xlabel("FPS ID")
plt.ylabel("Transaction Count")

# Show graphs
plt.tight_layout()
plt.show()
```
